## 결측치

```python
null_p = pd.read_csv('data/null_preprocessing.csv')
null_p = null_p.dropna().reset_index(drop=True)
null_p
```

```python
mean_col = null_p.variables[null_p.null_pre == -99].values
mean_col
```

```python
df[mean_col] = df[mean_col].fillna(df[mean_col].mean())
```

```python
df['F3TZYR2GPA'].isnull().sum()
```

```python
null_p[null_p.null_pre != -99]
```

```python
null_col = null_p[null_p.null_pre != -99].set_index('variables')['null_pre']
df = df.fillna(null_col)
```

```python
df.BYDOB_P.isnull().sum()
```

```python
df.BYDOB_P.describe()
```

```python
df['BYDOB_P'] = df['BYDOB_P'].fillna(df['BYDOB_P'].median())
df.BYDOB_P.isnull().sum()
```

```python
test = df[df.F3ONET2CURR==99].reset_index(drop=True)
train = df[df.F3ONET2CURR!=99].reset_index(drop=True)
```

```python
train.to_csv('data/train_ver03.csv', index=False)
test.to_csv('data/test_ver03.csv', index=False)
```
